```python
import csv

# Function to read and print values from 'bank_statement_expense.csv'
def read_expenses_file(expense_file):
    try:
        # Open the CSV file
        with open(expense_file, 'r', encoding='utf-8') as file:
            csv_reader = csv.DictReader(file)
            
            # Print each row in the CSV
            for row in csv_reader:
                print(row)
    except FileNotFoundError:
        print(f"Error: The file '{expense_file}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Main execution
if __name__ == "__main__":
    # Path to the bank_statement_expense.csv file
    expense_file = r'C:\Users\b8nguym23i\OneDrive - KUKA AG\Desktop\Python + SQL\Python New Handmade\Income Analytic\bank_statement_expenses.csv'
    
    # Call the function to read and print the file contents
    read_expenses_file(expense_file)
```

```python
def parse_amount(amount_str):
    return float(amount_str.replace(',', '.'))

def analyze_expenses(file_path):
    expenses = defaultdict(float)
    total_spent = 0
    start_date = None
    end_date = None

    with open(file_path, 'r', newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            date = datetime.strptime(row['Date'], '%Y%m%d')
            if start_date is None or date < start_date:
                start_date = date
            if end_date is None or date > end_date:
                end_date = date

            if row['Debit/credit'] == 'Debit':
                amount = parse_amount(row['Amount (EUR)'])
                category = categorize_expense(row['Name / Description'])
                expenses[category] += amount
                total_spent += amount

    return expenses, total_spent, start_date, end_date
def categorize_expense(description):
    # You can expand this function to categorize expenses more accurately
    if any(keyword in description.lower() for keyword in ['ah', 'albert heijn', 'jumbo', 'lidl', 'aldi']):
        return 'Groceries'
    
    elif any(keyword in description.lower() for keyword in ['restaurant', 'cafe', 'coffee', 'bakker']):
        return 'Dining out'
    
    elif 'gelmat' in description.lower():
        return 'Cash Withdrwal'
    
    elif 'ov' in description.lower():
        return 'Public Transport'
    else:
        return 'Other'
    
def print_expense_report(expenses, total_spent, start_date, end_date):
    print(f"Expense Report ({start_date.date()} to {end_date.date()}):")
    print("-" * 40)
    for category, amount in sorted(expenses.items(), key=lambda x: x[1], reverse=True):
        print(f"{category:<20} {amount:.2f}")
    print("-" * 40)
    print(f"Total Spent:          {total_spent:.2f}")
    
    days = (end_date - start_date).days + 1
    daily_average = total_spent / days
    print(f"Daily Average:        {daily_average:.2f}")

# Usage
file_path = 'bank_statement_expenses.csv'  # Replace with your CSV file path
expenses, total_spent, start_date, end_date = analyze_expenses(file_path)
print_expense_report(expenses, total_spent, start_date, end_date)
```
